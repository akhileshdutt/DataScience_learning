## Create Test Datasets

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# n_samples = sample points of 100 datapoints
# centres = mean clusters 
# cluster_std = standar deviation of each cluster ,lower means closer to each other, higher value means far away from each other
# n_features = sets the no of features or dimensions of the datasets
x,y = make_blobs(n_samples=100,centers=4,cluster_std=1,n_features=3)
plt.scatter(x[:,0],x[:,1],color="r",s=40)
plt.xlabel("dataset")
plt.show()
```

```python
from sklearn.datasets import make_moons
x,y = make_moons(n_samples=1000,noise=0.1)
plt.scatter(x[:,0],x[:,1],color="g",s=40)
plt.xlabel("dataset")
plt.show()
```

```python
from sklearn.datasets import make_circles
x,y = make_circles(n_samples=1000,noise=0.1)
plt.scatter(x[:,0],x[:,1],color="g",s=40)
plt.xlabel("dataset")
plt.show()
```

## Generate Test Datasests

```python
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples=100,centers=4,cluster_std=1,n_features=3)
plt.scatter(x[:,0],x[:,1],c=y,s=40)
plt.xlabel("dataset")
plt.show()
```

```python
#import the necessary libraries
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

# generate 2d classification dataset
X, y = make_classification(n_samples = 200, 
						n_features=16,
						n_redundant=1,
						n_informative=4,
						n_repeated=0,
						n_classes =10,
						n_clusters_per_class=1)

# Plot the generated datasets
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

Parameters:
n_samples=100: Specifies the number of data points (or samples) to generate. In this case, 100 samples will be created.

n_features=2: Sets the total number of features (or input variables) for each sample. Each data point will have 2 features.

n_redundant=0: This defines the number of redundant features, which are linear combinations of the informative features. Here, no redundant features are added (n_redundant=0).

n_informative=2: Specifies the number of informative features, meaning the features that are actually useful for predicting the target class. Since n_informative=2, both features will contribute useful information for class separation.

n_repeated=0: Determines the number of repeated features (duplicate features). Since it's set to 0, no features are repeated.

n_classes=3: Specifies the number of distinct classes (or labels) in the target variable. Here, there are 3 classes.

n_clusters_per_class=1: This defines how many clusters of data points will be generated for each class. Since n_clusters_per_class=1, each class will be represented by a single cluster.
