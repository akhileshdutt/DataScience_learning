```python
import matplotlib.pyplot as plt
```

```python
months = ["january","february","march","april","may"]
traffic_cong = [122,233,100,44,600]
```

```python
##create a line plot
plt.figure(figsize=(10,10))
plt.plot(months,traffic_cong)
plt.xlabel("Months ------>")
plt.ylabel("Traffic congestions ------>")
plt.show()
```

## Adding styles to the graph

```python
plt.plot(months,traffic_cong,marker="o",linestyle="dashed")
plt.xlabel("Months ------>")
plt.ylabel("Traffic congestions ------>")
plt.grid()
plt.show()
```

```python
plt.plot(months,traffic_cong,marker="o",linestyle="dashed",color="r")
plt.xlabel("Months ------>")
plt.ylabel("Traffic congestions ------>")
plt.grid()
plt.style.use('dark_background') ## the style commands should be before the show
plt.show()
```

We often have to adjust the plot size, right? And to do that, we need to use plt.figure(figsize=(x_size,y_size)) , Make sure to use this before the .plot command.

### Multiple Line Plot

```python
months = ["january","february","march","april","may"]
profit_1 =[23,425,367,83,56]
profit_2 = [245,256,578,124,165]
```

```python
plt.figure(figsize=(10,5))
#creating a line plot for product 1 profit with green line and circular markers
plt.plot(months,profit_1,marker="o",color="g",linestyle="-",label="Product_1")
#creating a line plot for product 2 profit with red line and square markers
plt.plot(months,profit_2,marker="s",color="r",linestyle="dashed",label="Product_2")
plt.xlabel("Months ------>")
plt.ylabel("Profits for different Products ----->")
plt.title("Profit generated in different months")
# in order to show different graphs for different products
plt.legend()
plt.style.use("default")
plt.show()
```

## Saving the plot

```python
plt.figure(figsize=(10,5))
#creating a line plot for product 1 profit with green line and circular markers
plt.plot(months,profit_1,marker="o",color="g",linestyle="-",label="Product_1")
#creating a line plot for product 2 profit with red line and square markers
plt.plot(months,profit_2,marker="s",color="r",linestyle="dashed",label="Product_2")
plt.xlabel("Months ------>")
plt.ylabel("Profits for different Products ----->")
plt.title("Profit generated in different months")
# in order to show different graphs for different products
plt.legend()
plt.style.use("default")
plt.savefig("MultiLinePlot.png")
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
import numpy as np

%matplotlib notebook
# Create a figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)

# Initialize the point to be animated
point, = ax.plot([], [], 'bo', markersize=10)

# Function to initialize the plot
def init():
    point.set_data([], [])
    return point,

# Function to update the animation
def animate(frame):
    x = np.linspace(0, 2 * np.pi, 1000)
    y = np.sin(x + 0.1 * frame)  # Vary the phase to create animation
    point.set_data(x, y)
    return point,

# Create the animation
ani = FuncAnimation(fig, animate, frames=100, init_func=init, blit=True, interval=50)

plt.title('Animated Point on Sine Wave')
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.grid(True)

plt.show()
```
