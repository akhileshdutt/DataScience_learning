## Categorical Plots

Categorical plots in data visualization are used to visualize categorical data, which consists of discrete and distinct categories or groups.

#### Count Plot
The Count Plot displays the counts of observations in each category. Suitable for visualizing the frequency or distribution of categorical data.

```python
plt.figure(figsize=(10,10))
sns.countplot(x="size",data=tips,palette="Blues_d") 
plt.title(" A basic Count Plot of total tips")
plt.show()
# Specify a specific name for a continuous color palette. --If you add "_d" to the name,
# it will be dark. (Light if not attached) --If you add "_r" to the name, the order will be reverse
```

### Swarm Plot

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x="size",y="tip",data=tips,palette="viridis",hue="size") 
plt.title(" A basic Count Plot of total tips")
plt.legend()
plt.show()
```

### Point Plots

```python
plt.figure(figsize=(8, 5))
sns.pointplot(x="day", y="tip", data=tips, palette="pastel")
plt.title("Average Tip by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip ($)")
plt.show()
```

### Box Plot

```python
plt.figure(figsize=(10,8))
sns.boxplot(x="smoker",y="tip",data=tips,fliersize=6,boxprops={"facecolor":"darkgreen"},flierprops={"marker":"o","markerfacecolor":"yellow"},whis=2.5)
plt.title("Smoking activity in the restaurant on day to day basis")
plt.xlabel("day")
plt.ylabel("smoker")
plt.show()
```

### Categorical Violin Plot

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x="day", y="tip", data=tips, palette="pastel",alpha=1.5)
plt.title("Average Tip by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip ($)")
plt.show()
```

### Cat Plot

```python
plt.figure(figsize=(8, 5))
sns.catplot(x="day", data=tips, palette="pastel",kind="count")
plt.title("Average Tip by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip ($)")
plt.show()
```

```python
plt.figure(figsize=(10, 20))
sn = sns.catplot(x="day", y="total_bill", data=tips,col="sex", palette="pastel",kind="bar",alpha=1)
sn.set_axis_labels("day","total_bill")
sn.set_titles("{col_name} {col_var}")
sn.set_xticklabels(["Thursday","Friday","Saturday","Sunday"])
plt.show()
```

## Uinvariate Plots

### KDE Plot(Kernel Density Estimation)

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=tips["tip"],fill=True,color="green",alpha=0.25)
plt.title("Average Tip by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip ($)")
plt.show()
```
