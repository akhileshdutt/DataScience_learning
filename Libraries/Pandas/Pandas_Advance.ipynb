### Use of iloc and loc

```python
df[['PassengerId','Survived','Pclass']].iloc[0:2]
```

```python
df.iloc[0:2,['PassengerId','Survived','Pclass']]	
## iloc uses the inbuilt indexing in the pandas dataframe that means in pandas dataframe the given column names have values as 0,1,2
```

```python
#solution to the iloc problem
df.iloc[0:2,[0,1,2]]	
```

```python
#if we don't want to use iloc we can use loc
df.loc[0:2,['PassengerId','Survived','Pclass']]	
```
