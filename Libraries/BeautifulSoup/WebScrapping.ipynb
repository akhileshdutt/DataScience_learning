#### find gives us string and find_all returns us ResultSet

```python
data1=[]
for i in soup.find_all("div",class_="quote"):
    sp = i.find("span",class_="text").text[1:-1]
    auth = i.find("small",class_="author").text
    link = i.find("a").get("href")
    tags=[]
    for tag in i.find_all("a",class_="tag"):
        tags.append(tag.text)
    tags = ",".join(tags)
    data1.append([sp,auth,link,tags])
```

```python
data1
```

```python
import pandas as pd
df = pd.DataFrame(data1,columns=["Quotes","Author","Link","Tags"])
```

```python
df
```

```python
df.to_html("Quotes.html",justify="left")
```

```python
df.to_csv("Quotes1.csv",index=False)
```

### automating the process (not applicable to all the websites here we had a pattern)

```python
data=[]
for page in range(2,11):
    link ="https://quotes.toscrape.com/page/"+str(page)
    res = requests.get(link)
    soup = BeautifulSoup(res.text,"html.parser")
    for i in soup.find_all("div",class_="quote"):
        sp = i.find("span",class_="text").text[1:-1]
        auth = i.find("small",class_="author").text
        link = i.find("a").get("href")
        tags=[]
        for tag in i.find_all("a",class_="tag"):
            tags.append(tag.text)
        tags = ",".join(tags)
        data.append([sp,auth,link,tags])    
```

```python
data
```

```python
df1=pd.DataFrame(data,columns=["Quotes","Author","Link","Tags"])
```

```python
df2 = pd.concat([df,df1])
```

```python
df2.to_csv("All_Quotes.csv",index=False)
```
