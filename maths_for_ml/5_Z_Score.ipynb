```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pylab
```

## Implementation of Z-scores from scratch

```python
values = np.array([10,4,25,67,24,26,78,98,45,36,68,56])
mu = values.mean()
sigma = values.std()
z_scores = (values - mu)/sigma
for i, (value, z_score) in enumerate(zip(values, z_scores)):
    print(f"Data point {i+1}: {value}, Z-score: {z_score:.4f}")

plt.plot(z_scores)
```

## InBuilt function of Z-scores

```python
from scipy.stats import zscore
values = np.array([10,4,25,67,24,26,78,98,45,36,68,56])
z_scores = stats.zscore(values)
for i, (value,z_score) in enumerate(zip(values, z_scores)):
    print(f"Data point {i+1}: {value}, Z-score: {z_score:.4f}")

plt.plot(z_scores)
```

#### calculate the probability that an IQ is lower than 85 given the average IQ in India is 100 and the standard deviation is 15, 

```python
from scipy.stats import norm
mu = 100
sigma = 15
given_iq = 85
z_score = (given_iq - mu)/sigma
prob_lower_than_85 = norm.cdf(z_score)
print(f"Probability in percenatge: {prob_lower_than_85*100 :.1f}%")
```
